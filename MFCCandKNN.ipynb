```python
import numpy as np
import sounddevice as sd
import matplotlib.pyplot as plt
import time
```

```python
import os
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 20,25
```

```python
fs = 44100
```

```python
duration = 10  # seconds
lst = []
for i in range(3):
    print("Recording in 1...")
    time.sleep(1)
    os.system('spd-say "Speak"')
    time.sleep(0.5)
    current_recording = sd.rec(int(duration * fs), samplerate=fs, channels=1, dtype=np.float64)
    time.sleep(10)
    lst.append(current_recording)
    input("Press ENTER")
```

```python
duration = 10  # seconds
recordings = []
#recordings = np.zeros((2,3,fs*3,1))
for i in range(25):
    print(i+1)
    print("Recording in 1 second...")
    time.sleep(1)
    os.system('spd-say "Speak"')
    time.sleep(0.5)
    current_recording = sd.rec(int(duration * fs), samplerate=fs, channels=1, dtype=np.float64)
    time.sleep(10)
    recordings.append(current_recording)
    input("Press ENTER")
print("Finished.")
```
